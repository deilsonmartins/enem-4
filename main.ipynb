```
import numpy as np
import pandas as pd
from sklearn import metrics

df_train = pd.read_csv('gdrive/My Drive/codenation/Modulos/enem-4/train.csv', index_col=0)
df_test = pd.read_csv('gdrive/My Drive/codenation/Modulos/enem-4/test.csv')
```

```
df_train
```

```
colunas_uteis = ['IN_TREINEIRO', 'NU_IDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU']
```

```
df = df_train[colunas_uteis]
df.head()
```

```
df.info()
```

```
df.isna().sum()
```

```
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```
AlunosTreineiros = (df['IN_TREINEIRO'] == 1).sum()

AlunosNaoTreineiros = (df['IN_TREINEIRO'] == 0).sum()

percentual = round(100 * AlunosTreineiros/AlunosNaoTreineiros, 3)

print("Alunos treineiros: " + str(AlunosTreineiros ))
print("Alunos não treineiros: " + str(AlunosNaoTreineiros))
print("Percentual de alunos treineiros " + str(percentual) + "%")
```

```
X = df.drop(['IN_TREINEIRO'], axis=1)
y = df['IN_TREINEIRO']
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)

print("X_train " + str(X_train.shape))
print("X_test " + str(X_test.shape))
print("y_train " + str(y_train.shape))
print("y_test " + str(y_test.shape))
```

```
AlunosTreineiros = (y_train == 1).sum()

AlunosNaoTreineiros = (y_train == 0).sum()

percentual = round(100 * AlunosTreineiros/AlunosNaoTreineiros, 3)

print("########### TREINO ########")
print("Alunos treineiros: " + str(AlunosTreineiros ))
print("Alunos não treineiros: " + str(AlunosNaoTreineiros))
print("Percentual de alunos treineiros " + str(percentual) + "%")
```

```
AlunosTreineiros = (y_test == 1).sum()

AlunosNaoTreineiros = (y_test == 0).sum()

percentual = round(100 * AlunosTreineiros/AlunosNaoTreineiros, 3)

print("########### TESTE ########")
print("Alunos treineiros: " + str(AlunosTreineiros ))
print("Alunos não treineiros: " + str(AlunosNaoTreineiros))
print("Percentual de alunos treineiros " + str(percentual) + "%")
```

```
from imblearn.ensemble import BalancedBaggingClassifier

bbc = BalancedBaggingClassifier(random_state=42)

bbc.fit(X_train, y_train)

y_pred = bbc.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

```
import numpy as np
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred)
```
